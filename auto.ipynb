A stroke is when the piston moves from the top center to the bottom of the cylinder. During a combustion cycle, while the piston moves down the cylinder, it captures air and gas. As the piston moves back up to the top, the exhaust valve opens to expel the exhaust.

```python
sns.boxplot(data = auto, x = 'stroke')

#here we can see outliers
```

The compression ratio (CR) is defined as the ratio of the volume of the cylinder and its head space (including the pre-combustion chamber, if present) when the piston is at the bottom of its stroke to the volume of the head space when the piston is at the top of its travel ('top dead centre', tdc).

```python
sns.boxplot(data = auto, x = 'compression-ratio')

#here we ca se outliers
```
